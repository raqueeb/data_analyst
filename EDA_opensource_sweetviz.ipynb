```
import sweetviz as sv
import pandas as pd
train_df = pd.read_csv('https://github.com/raqueeb/mltraining/raw/master/ML-workbook/train.csv')
test_df = pd.read_csv('https://github.com/raqueeb/mltraining/raw/master/ML-workbook/test.csv')
```

### দুটো তুলনা --- ট্রেনিং এবং টেস্ট সেটের মধ্যে

১. রিপোর্ট জেনারেট করি

```
# টার্গেট ফিচার "Survived" কে ধরে

comparison_report = sv.compare([train_df,'Train'], [test_df,'Test'], target_feat='Survived')
```

### রিপোর্ট তৈরী করছি, স্ট্যাটিক ফাইল দেখা যায় নিজের পিসিতে 

```
comparison_report.show_html()
```

### রিপোর্ট ডাউনলোড অথবা নোটবুকে দেখতে পারি

```
comparison_report.show_notebook() # ডিফল্ট ভ্যালু (w="100%", h=750, layout="vertical")
```

### ভার্টিকাল অথবা ওয়াইডস্ক্রিন লেআউট

```
comparison_report.show_notebook(layout='widescreen', w=1920, h=1080, scale=0.9) 
```
